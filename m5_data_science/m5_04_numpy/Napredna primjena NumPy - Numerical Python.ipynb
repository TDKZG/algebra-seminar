```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline

import datetime
import json
import requests

unix_dates = [
    1683565200, 
    1683478800,
    1683392400, 
    1683306000,
    1683219600
]

cities = {
    'Pula': [0, 44.86833, 13.84806],
    'Pazin': [30, 45.23878, 13.93497],
    'Motovun': [25, 45.33673, 13.82821],
    'Labin': [4, 45.09485, 14.12319],
    'Buzet': [40, 45.40788, 13.96559],
    'Buje': [12, 45.40837, 13.65914],
    'Visnjan': [10, 45.27602, 13.71887],
    'Kanfanar': [17, 45.12284, 13.83850],
    'Zminj': [23, 45.14391, 13.90868],
    'Vodnjan': [7, 44.95896, 13.85134]
}
```

```python
def filter_city_weather_data_per_hour(hourly_data, city_name):
    temp = round(hourly_data["temp"] - 273.15, 2)
    humidity = hourly_data["humidity"]
    pressure = hourly_data["pressure"]
    description = hourly_data["weather"][0]["description"]
    dt = hourly_data["dt"]
    wind_speed = hourly_data["wind_speed"]
    wind_deg = hourly_data["wind_deg"]
    dt_object = datetime.datetime.fromtimestamp(dt)
    formatter = "%Y-%m-%d %H:%M:%S"
    dt_formatted = datetime.datetime.strftime(dt_object, formatter)
    return [temp, humidity, pressure, description, dt, wind_speed, wind_deg, city_name, dt_formatted]
```

```python
api_url = "http://api.openweathermap.org/data/2.5/onecall/timemachine"

for city, values in cities.items():
    data_frame_list = []
    for day in unix_dates:
        city_res = json.loads(requests.get(f"{api_url}?lat={values[1]}&lon={values[2]}&dt={day}&appid={api_key}").text)
        
        for i in range(len(city_res["hourly"])):
            df_row = filter_city_weather_data_per_hour(city_res["hourly"][i], city)
            df_row.append(values[0])
            
            data_frame_list.append(df_row)
    
    headings = ["temp", "humidity", "pressure", "description", "dt", "wind_speed", "wind_deg", "city", "day", "distance"]
    city_df = pd.DataFrame(data_frame_list, columns=headings)
    city_df.to_csv(f"./datoteke/df_{city}.csv")
```

```python
def display_rose_wind(values, city_name, max_value):
    N = 8
    theta = np.arange(0., 2 * np.pi, 2 * np.pi / N)
    radii = np.array(values)
    plt.axes([0.025, 0.025, 0.95, 0.95], polar=True)
    colors = [(1 - x/max_value, 1-x/max_value, 0.75) for x in radii]
    plt.bar(theta + np.pi/8, radii, width=(2*np.pi/8), bottom=0.0, color=colors)
    plt.title(city_name, x=0.2, fontsize=20)
```

```python
hist, bins = np.histogram(df_pula["wind_deg"], 8, [0, 360])
display_rose_wind(hist, "Pula", max(hist))
```
