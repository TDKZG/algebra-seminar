```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_movies = pd.read_csv("../Data/Filmovi/movie_dataset.csv", index_col=0)
df_movies
```

```python
df_movies.info()
```

```python
df_movies.budget.describe()
```

```python
df_movies.popularity.describe()
```

```python
df_budget_popularity = df_movies[["budget", "popularity"]]
df_budget_popularity.head()
```

```python
df_budget_popularity.describe()
```

```python
df_budget_popularity.plot(subplots=True)
```

```python
df_budget_popularity.iloc[::10].plot(style="o", markersize=1, subplots=True)
```

```python
df_budget_popularity.budget.plot(kind="hist", bins=100)
```

```python
df_budget_popularity.popularity.plot(kind="hist", bins=50)
```

```python
df_budget_popularity.plot(kind="box", subplots=True)
```

```python
df_budget_popularity.plot(
    kind="scatter",
    x="budget",
    y="popularity",
    alpha=0.5
)
```

```python
df_no_outliers = df_budget_popularity[(df_budget_popularity.popularity < 50) & (df_budget_popularity.budget < 8e7)]
df_no_outliers.plot(kind="box", subplots=True)
```

```python
df_no_outliers.iloc[::5].plot(
    kind="scatter",
    x="budget",
    y="popularity",
    alpha=1
)
```
