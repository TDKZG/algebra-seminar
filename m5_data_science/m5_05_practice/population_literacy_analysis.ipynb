```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path_15_24 = "literacy_rate_youth_total_percent_of_people_ages_15_24.csv"
path_15_above = "literacy_rate_adult_total_percent_of_people_ages_15_and_above.csv"
path_out_of_school = "out-of-school-children-of-primary-school-age-by-world-region.csv"
```

```python
dir_path = "../Data/Pismenost stanovnistva"
```

```python
df_literacy_15_24 = pd.read_csv(f"{dir_path}/{path_15_24}")
df_literacy_15_24
```

```python
df_literacy_15_above = pd.read_csv(f"{dir_path}/{path_15_above}")
df_literacy_15_above
```

```python
df_out_of_school = pd.read_csv(f"{dir_path}/{path_out_of_school}")
df_out_of_school
```

```python
kolone = list(df_out_of_school.columns.values)
kolone[0] = 'country'
kolone[2] = 'year'
df_out_of_school.columns = kolone
df_out_of_school.columns
```

```python
common_countries = np.intersect1d(df_literacy_15_24.country.unique(), 
                                  df_literacy_15_above.country.unique(), assume_unique=True)
common_countries = np.intersect1d(common_countries, df_out_of_school.country.unique())
len(common_countries)
```

```python
df_literacy_15_24 = df_literacy_15_24[df_literacy_15_24.country.isin(common_countries)]
df_literacy_15_above = df_literacy_15_above[df_literacy_15_above.country.isin(common_countries)]
df_out_of_school = df_out_of_school[df_out_of_school.country.isin(common_countries)]
```

```python
df_literacy_15_24
```

```python
df_literacy_15_24 = pd.melt(
    df_literacy_15_24,
    id_vars = ["country"],
    value_vars = [str(year) for year in range(1975, 2012)],
    var_name = "year", 
    value_name = "literacy_15_24"
)
df_literacy_15_24
```

```python
df_literacy_15_24[df_literacy_15_24.country == "Aruba"]
```

```python
df_literacy_15_above = pd.melt(
    df_literacy_15_above,
    id_vars = ["country"],
    value_vars = [str(year) for year in range(1975, 2012)],
    var_name = "year", 
    value_name = "literacy_15_above"
)
df_literacy_15_above
```

```python
df_merged = pd.merge(df_literacy_15_24, df_literacy_15_above, how="inner", on=["country", "year"])
df_merged
```

```python
def display_data_for_country(df, country):
    okvir, axes = plt.subplots(1, 2, figsize=(15, 5))

    df[df.country == country].plot(
        ax=axes[0],
        kind='bar',
        x='year',
        y=['literacy_15_24', 'literacy_15_above'],
        color=['red', 'green'],
        legend=True
    )

    df[df.country == country].plot(
        ax=axes[1],
        kind='line',
        x='year',
        y='Out-of-school children of primary school age, both sexes (number)',
        legend=True
    )
```
