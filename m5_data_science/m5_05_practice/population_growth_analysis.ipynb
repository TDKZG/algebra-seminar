Sada imamo iste godine i drzave za sve tablice. Idemo ih presloziti da godine budu u jednom stupcu.

```python
def melt_dataframe(df, value_name):
    return pd.melt(
        df,
        id_vars=["country"],
        value_vars=common_columns.tolist()[:-1],
        var_name="year",
        value_name=value_name
    )
```

```python
population_total_df = melt_dataframe(population_total_df, "population_total")
population_total_df
```

```python
population_density_df = melt_dataframe(population_density_df, "population_density_sq_km")
population_density_df
```

```python
agricultural_df = melt_dataframe(agricultural_df, "agricultural_land_%")
agricultural_df
```

```python
forest_area_df = melt_dataframe(forest_area_df, "forest_area_sq_km")
forest_area_df
```

```python
surface_area_df = melt_dataframe(surface_area_df, "surface_area_sq_km")
surface_area_df
```

###### spajanje svih 5 dataframe-ova u jedan
- prva dva stupca su identicna u svim dataframe-ovima jer smo uzeli presjek tih stupaca
- razlikuju se po trecem stupcu
- trebamo dobiti finalni dataframe koji ima 5157 redaka i 7 stupaca

```python
merged_df = pd.merge(population_total_df, population_density_df, how="inner", on=["country", "year"])
merged_df = pd.merge(merged_df, agricultural_df, how="inner", on=["country", "year"])
merged_df = pd.merge(merged_df, forest_area_df, how="inner", on=["country", "year"])
merged_df = pd.merge(merged_df, surface_area_df, how="inner", on=["country", "year"])
merged_df
```

```python
merged_df.info()
```

```python
merged_df = merged_df.sort_values(["country", "year"]).fillna(method="bfill")
merged_df
```

```python
merged_df[merged_df.country == "Ethiopia"]
```

```python
merged_df.info()
```

```python
interesting_columns = ["population_total", "population_density_sq_km", "forest_area_sq_km", "surface_area_sq_km"]
unique_chars = set()

for ind, row in merged_df[interesting_columns].iterrows():
    for value in row:
        unique_chars |= set(value)
            
unique_chars
```

```python
interesting_columns = ["population_total", "population_density_sq_km", "forest_area_sq_km", "surface_area_sq_km"]
unique_chars = set()

for ind, row in merged_df[interesting_columns].iterrows():
    unique_chars.update({char for value in row for char in value})
            
unique_chars
```

```python
def convert(x):
    result = None
    
    if "B" in x:
        result = float(x.replace("B", "")) * 1_000_000_000
    elif "M" in x:
        result = float(x.replace("M", "")) * 1_000_000
    elif "k" in x:
        result = float(x.replace("k", "")) * 1_000
    else:
        result = float(x)
    
    return result
```

```python
merged_df[interesting_columns] = merged_df[interesting_columns].applymap(convert)
merged_df
```

```python
merged_df["available_area_sq_km"] = merged_df["surface_area_sq_km"] * (1 - merged_df["agricultural_land_%"]/100) - merged_df["forest_area_sq_km"]
merged_df
```

```python
merged_df["population_density_available_area_sq_km"] = merged_df["population_total"] / merged_df["available_area_sq_km"]
merged_df
```

```python
merged_df["population_density_available_area_sq_km"] = merged_df["population_density_available_area_sq_km"].round(2)
merged_df
```

```python
merged_df[merged_df.country == "Angola"]
```

```python
def display_data_for_country(country):
    fig, axes = plt.subplots(2, 3, figsize=(15, 5))
    df = merged_df[merged_df.country == country]
    
    df.plot(
        ax=axes[0, 0],
        kind="line",
        x="year",
        y="population_total",
        color="red",
        legend=True
    )
    
    df.plot(
        ax=axes[0, 1],
        kind="line",
        x="year",
        y="population_density_sq_km",
        color="purple",
        legend=True
    )
    
    df.plot(
        ax=axes[0, 2],
        kind="line",
        x="year",
        y="population_density_available_area_sq_km",
        color="yellow",
        legend=True
    )
    
    df.plot(
        ax=axes[1, 0],
        kind="line",
        x="year",
        y="agricultural_land_%",
        color="brown",
        legend=True
    )
    
    df.plot(
        ax=axes[1, 1],
        kind="line",
        x="year",
        y="forest_area_sq_km",
        color="green",
        legend=True
    )
    
    df.plot(
        ax=axes[1, 2],
        kind="line",
        x="year",
        y="available_area_sq_km",
        color="blue",
        legend=True
    )
```
