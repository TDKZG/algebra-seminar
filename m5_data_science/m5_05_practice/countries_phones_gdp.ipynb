```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_computers = pd.read_csv("../Data/Trziste racunala i pametnih telefona/personal_computers_total.csv")
df_computers
```

```python
df_computers = df_computers.set_index("country").transpose()
df_computers
```

```python
df_computers.index = pd.to_datetime(df_computers.index)
df_computers
```

```python
df_computers.info()
```

```python
df_computers.replace({"k": "000", "M": "000000"}, regex=True)
```

```python
def eval_numbers(str_num):
    if isinstance(str_num, float):
        return str_num
    return eval(str_num)
```

```python
df_computers = df_computers.replace({"k": "*1e3", "M": "*1e6"}, regex=True).applymap(eval_numbers)
df_computers
```

```python
df_no_nan = df_computers.iloc[-5:].dropna(axis=1).astype(int)
df_no_nan
```

```python
(df_no_nan.iloc[-1] - df_no_nan.iloc[0]) / df_no_nan.iloc[0]
```
